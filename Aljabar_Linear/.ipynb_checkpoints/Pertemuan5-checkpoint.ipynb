<h3>Aljabar Matriks</h3>

Hukum Penjumlahan pada matriks :

1. Hukum Komutatif : **$A + B = B + A$**

2. Hukum Asosiatif : **$A + (B+C) = (A+B) + C$**

3. Hukum asosiatif perkalian : **$A(BC) = (AB)C$**

4. Hukum Distributif kiri : **$A(B+C) = AB + BC$**

5. Hukum Distributif Kanan : **$(B + C) A = BA + BC$**

6. Hukum Distributif penskalaan : **$a(B + C) = aB + aC$**

7. Hukum Distributif penskalaan lainnya : **$(a + b)C = aC + bC)$**

8. Hukum asosiatif penskalaan : **$a(bC) = (ab)C$**

9. Hukum komutatif penskalaan : **$a(BC) = (aB)C = B(aC)$**


Kegagalan Pencoretam : Dalam matrix jika ab = ac maka tidak bisa mencoret a kemudian menghukumi bahwa b = c

Tranposisi, atau biasa dilambangkan dengan T = adalah simbol yang menandakan bahwa matrix tersebut di balik, contoh : semisal ada matrikx A = m x n, maka jika di Tranpos menjadi A = n x m
