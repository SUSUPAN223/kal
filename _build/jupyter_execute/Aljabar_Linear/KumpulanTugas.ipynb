**SOAL SISTEM PERSAMAAN 3 VARIABEL :**

$$\begin{aligned}
x + y + z &= 10 \\
2x + y + z &= 14 \\
x + 2y + 3z &= 18
\end{aligned}$$

Dalam Matrix :

$$
\begin{bmatrix}
1&1&1&|&10\\
2&1&1&|&14\\
1&2&3&|&18
\end{bmatrix}
$$

**JAWABAN :**

- <i><b>A.add_multiple_of_row(1,0,-2)</b></i> = mengalikan -2 dengan seluruh baris 1, kemudian menambahkan hasilnya ditambahkan ke seluruh baris ke 2

$$
\begin{bmatrix}
1&1&1&|&10\\
0&-1&-1&|&-6\\
1&2&3&|&18
\end{bmatrix}
$$

- <i><b>A.add_multiple_of_row(2,0,-1)</b></i> = mengalikan -1 dengan seluruh baris ke 1, dan menambahkan hasilnya dengan seluruh baris ke 3

$$
\begin{bmatrix}
1&1&1&|&10\\
0&-1&-1&|&-6\\
0&1&2&|&8
\end{bmatrix}
$$

- <i><b>A.rescale_row(1,-1)</b></i> = mengalikan -1 dengan seluruh baris ke 2

$$
\begin{bmatrix}
1&1&1&|&10\\
0&1&1&|&6\\
0&1&2&|&8
\end{bmatrix}
$$

- <i><b>A.add_multiple_of_row(2,1,-1)</b></i> = mengalikan -1 dengan seluruh baris 2 kemudian menambahkan hasilnya ke seluruh baris 3

$$
\begin{bmatrix}
1&1&1&|&10\\
0&1&1&|&6\\
0&0&1&|&2
\end{bmatrix}
$$

- <i><b>A.add_multiple_of_row(1,2,-1)</b></i> = mengalikan -1 dengan seluruh baris ke 3, kemudian menambahkan hasilnya ke seluruh baris 2

$$
\begin{bmatrix}
1&1&1&|&10\\
0&1&0&|&4\\
0&0&1&|&2
\end{bmatrix}
$$

- <i><b>A.add_multiple_of_row(0,2,-1)</b></i> =  mengalikan -1 ke seluruh baris ke 3 dan menambahkan hasilnya ke seluruh baris ke 1

$$
\begin{bmatrix}
1&1&0&|&8\\
0&1&0&|&4\\
0&0&1&|&2
\end{bmatrix}
$$

- <i><b>A.add_multiple_of_row(0,1,-1)</b></i> = mengalikan -1 dengan seluruh baris ke 2 kemudian menambahkan hasilnya ke seluruh baris 1

$$
\begin{bmatrix}
1&0&0&|&4\\
0&1&0&|&4\\
0&0&1&|&2
\end{bmatrix}
$$

Maka, ditemukan hasil : $$\begin{aligned}x = 4, y = 4, z = 2\end{aligned}$$
