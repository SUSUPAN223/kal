Matriks adalah urutan(array) bilangan real yang di atur dengan caara yang sangat khusus

Matriks dinyatakan sama jika m x n dan m' x n', dengan kata lain memiliki natrix A = B jika dan hanya jika A dan B memiliki bentuk yang sama

Contoh kesamaan matrix :

$$
A =
\begin{bmatrix}
1&2&3&4\\
\end{bmatrix}
$$

$$
B = 
\begin{bmatrix}
1\\
2\\
3\\
4
\end{bmatrix}
$$
