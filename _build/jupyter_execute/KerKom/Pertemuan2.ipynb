AI sendiri memiliki hierarki dalam penempatannya, yakni:

1. Machine Learning : yakni mesin belajar dari data yang diberikan
2. Deeo Learning : yakni mesin belajar menggunakan ANN (Deep neural network)
3. Computational Intelligence : berisi Soft Computing, Optimasi numerik, Adaptasi, Ketidakpastian

**FUZZY LOGIC**

Fuzzy digunakan untuk mengukur suatu derajat, atau bisa disebut sistem kontrol, contoh:

1. Mesin Cuci : Misal bebannya ringan, dengan air yang sedikit maka sistem akan mengkontrol menggunakan daya yang sesuai, entah memerlukan kecepatan yang tinggi, sedang atau rendah itu adalah yang disebut defuzzification, setelah itu baru muncul output.
2. AC (Air Conditioner) : Pada AC, sistem menyesuaikan dengan tingkat kecepatan kipas dan suhu di dalam ruangan, jika suhu terdeteksi dalam keadaan tinggi, maka sistem akan otomatis mempercepat kipas untuk menurunkan suhu, dan jika sudah turun suhunya maka ia akan otomatis memperlambat proses pendinginan, bahkan ada yang mati untuk sementara tergantung dengan program AC yang ada, tujuannya supaya pengguna tidak lagi menggunakan ON/OFF manual.
