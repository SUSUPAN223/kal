**Mengapa Fuzzy Dibutuhkan?**

<p>Fuzzy adalah sebuah sistem yang mengajarkan kepada komputer makna dari ambiguitas, contoh : komputer atau sistem semestinya hanya mengenal kepastian yakni ya atau tidak, di ibaratkan pada tinggi badan manusia, pada dasarnya komputer hanya mengerti tinggi dan pendek, namun pada kenyataannya ada penilaian "agak tinggi" "lumayan tinggi" "tinggi" "sangat tinggi", kalimat kalimat tersebut adalah ambiguitas yang tidak bisa ditentukan hanya dengan tinggi dan pendek, maka fungsi dari fuzzy adalah mengajarkan sistem bahwa misal, jika seseorang itu "agak tinggi" maka penilaiannya bukan 1, melainkan 0,6, jika seseorang itu "lumayan tinggi" dia memiliki nilai 0,7 dst.</p>

Operasi Himpunannya ada 3 :

- Complement

- Union
